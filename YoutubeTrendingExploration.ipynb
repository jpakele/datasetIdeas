```python
from apiclient.discovery import build
```

```python
youtube = build('youtube', 'v3', developerKey = apiKey)
```

```python
type(youtube)
```

```python
from datetime import datetime
```

```python
startTime = datetime(year=2020, month = 1, day = 1).strftime('%Y-%m-%dT%H:%M:%SZ')
endTime = datetime(year=2020, month = 1, day = 8).strftime('%Y-%m-%dT%H:%M:%SZ')
startTime
```

```python
results = youtube.search().list(part = 'snippet',
                                maxResults= 1000,
                                type = 'video',
                                publishedAfter = startTime,
                                publishedBefore = endTime).execute()
```

```python
for item in sorted(results['items'], key=lambda x:x['snippet']['publishedAt']):
    print(item['snippet']['title'], item['snippet']['publishedAt'])
```
